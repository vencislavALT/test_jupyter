# 2. Snake

Да направим добрата стара змия.

По подадени брой редове и колони генерираме поле със същите размери.

Змията трябва да може да се придвижда в четирите посоки, при натискане на WASD. След всяко придвижване змията и полето се изпринтват, след което се чака ново натискане на WASD.

След като змията изяде ябълка, се генерира нова ябълка на свободно квадратче.

Ако змията се удари в ръба на полето - умираме и се изпринтва резултат, колко ябълки сме успели да изядем.

Да започнем със змия, която не расте след като изяде ябълката.

# 4. Roulette
Изградете прост начин за игра на рулетка. Ако не ви е позната, основите на играта са прости - имаме въртящ се кръг със сектори - числа (освен числа, секторите са един от три цвята - червен, черен и специалният зелен). По принцип има много начини за залагане на рулетка, но нека за момента да се съсредоточим върху това да напишем програма, която позволява един играч да заложи (console input) на числата от 1 до 18, червено, черно, четно и нечетно.  
За момента няма да изчисляваме конкретните цифри за залозите!  
След това завъртаме рулетката, виждаме кое число се е паднало и преценяваме дали е спечелил играча.
